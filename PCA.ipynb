```python

import numpy as np
from numpy.linalg import eig

```

```python
A=np.random.random((784,20))
#p20 samples of mnist
C=np.cov(A)
print(C)
```

```python
w,v=eig(C)
print(w)
```

```python
s=[]
for i in range(784):
  s.append(abs(w[i]))
s.sort(reverse=True)
Sum=sum(s)
sumval=0
counter=0
for i in range (784) :
  sumval=sumval+s[i]
  if sumval/Sum >0.98:
    counter=counter+1
    break
  else:
    counter=counter+1
print(Sum)
print(sumval)
print(counter)
```

we conclude that we just need about 20 features from 784 feaures of mnist datasets to store at least 98 percents of features
